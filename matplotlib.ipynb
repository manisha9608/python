```python
import numpy as np;
import matplotlib.pyplot as plt;
```

```python
x = np.array([4,8,16,32,64])
```

```python
y_random = np.array([0.414, 0.557, 0.697, 0.8502, 0.9611])
y_fifo = np.array([0.397, 0.544, 0.689, 0.845, 0.957])
y_lru = np.array([0.381, 0.537, 0.6874, 0.857, 0.9578])
```

```python
scalingFactor = 100
plt.plot(x, y_random*scalingFactor, c='#0165fc', label='Random')
plt.plot(x, y_fifo*scalingFactor, c='#ec2d01', label='FIFO')
plt.plot(x, y_lru*scalingFactor, c='#23c48b', label='LRU')
plt.legend()
plt.grid(True)
plt.xlabel('Cache Size')
plt.ylabel('Hit Ratio (%)')
```
