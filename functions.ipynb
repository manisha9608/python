# Functions

```python
def calcInterest(principal, rate, time):
    print('Param values', principal, rate, time)
    return (principal*rate*time)/100
calcInterest(100,10,10)
```

```python
calcInterest(rate =5, principal =500, time = 2)
```

```python
calcInterest(rate =5, principal =500, 2)
```

## Multiple Arguments

```python
def fun2(*nums):
    print('Input', nums)
    print (type(nums))
    for i in nums:
        print(i, end = ',')
    print()

```

```python
fun2(2,3)
```

```python
fun2([2,3,4])
```

## Name Space

```python
a = 5

def fun(b):
    print('Local value', b)
    global a
    print('Global value', a)
    a = 3
```

```python
print('Before', a)
fun(2)
print('Outside', a)
```

## shallow copying

```python
def fun3(a):
    print(a)
    a = a*2
```

```python
x = 2
fun3(x) #Immutable object, therefore we are fine here
print(x)
```

```python
def fun4(list):
    print(list)
    list[0]=2*list[0]
    
```

```python
x = [2,3,4]
fun4(x)
print(x)
```

## Lambda function

```python
# Example1
a = lambda x: x*x
```

```python
def fun4(list, fun):
    return [fun(x) for x in list]
```

```python
x = [2,3,4]
print(fun4(x, a))
```

```python
# Example2
# Sort on the basis of count of unique letters inn the name
cities = ['hyderabad', 'mumbai', 'mississippi', 'varanasi', 'gaya', 'bengaluru']
#.        7             5.         4.             6.         3.       7
```

```python
cities.sort(key = lambda x: len(set(x)))
print(cities)
```

```python
# Example3
# Sort list onn the basis of second element of tuple
list1 = [(1,3,2), (3,6,9), (1,4,7), (4,1,5)]
list1.sort(key = lambda tuple: tuple[1])
print(list1)
```

```python
# Example4
# Filter function
a = [77,45,32,93,2,0,55,88]
x = filter(lambda i: (i%2 == 0), a)
print(list(x))
```
